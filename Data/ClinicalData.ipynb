## Importing clinical data

```python
dataFile = "clinical.tsv"

#dataFrame 
df = pd.read_csv(filepath_or_buffer=dataFile, sep='\t', nrows=0)

#https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html#io-chunking
with pd.read_csv(filepath_or_buffer=dataFile, sep='\t', chunksize=100, na_values='\'--') as reader:
    for chunk in reader:
        df=pd.concat([df, chunk])
```
